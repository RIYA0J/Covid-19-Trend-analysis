```python

countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rates
```

```python
happiness_report_csv = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report_csv.head()
```

```python
# @title Overall rank vs GDP per capita

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Overall rank']
  ys = series['GDP per capita']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = happiness_report_csv.sort_values('Overall rank', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Overall rank')
_ = plt.ylabel('GDP per capita')
```
